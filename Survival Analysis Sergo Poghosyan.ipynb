```python
import lifelines
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from lifelines import WeibullFitter, ExponentialFitter, LogNormalFitter, LogLogisticFitter, WeibullAFTFitter, LogNormalAFTFitter, LogLogisticAFTFitter, PiecewiseExponentialRegressionFitter
```

```python
telco = pd.read_csv('telco.csv') 
telco.head()
```

We need to transform several columns into categorical variables by using encodings. These columns include make region, marital status, education level, gender, retirement status, voice type, internet usage, forwarding preference, and customer satisfaction category.

```python
cols = ['region','marital','ed','gender', 'retire','voice','internet','forward','churn', 'custcat']
 
encoded = pd.get_dummies(telco,
               columns=cols,
               prefix=cols,
               drop_first=False)
 
encoded.head()
```

```python
telco['region'].unique()
```


We've identified a redundant column in our dataset that could cause multicollinearity issues in our model. We'll retain only the necessary columns for our further analysis, discarding the redundant one.

```python
encoded.columns
```

```python
survival = encoded[["age", "income", "churn_Yes","tenure","marital_Married",
                    "forward_Yes","internet_Yes", "custcat_E-service",
                          "custcat_Plus service","custcat_Total service",
                    "region_Zone 3","region_Zone 2"]]
survival.head()
```

Accelerated Failure Time Model

```python
survival.info()
```

```python
wb = WeibullFitter()
ex = ExponentialFitter()
log = LogNormalFitter()
loglogis = LogLogisticFitter()

we_aft = WeibullAFTFitter()
we_aft.fit(survival, duration_col='tenure', event_col='churn_Yes')
we_aft_pd = we_aft.predict_survival_function(survival).transpose()
avg_we = we_aft_pd.mean().to_list()
we_aft_pd
```

```python
logn_aft = LogNormalAFTFitter()
logn_aft.fit(survival, duration_col='tenure', event_col='churn_Yes')
logn_aft_pd = logn_aft.predict_survival_function(survival).transpose()
avg_logn = logn_aft_pd.mean().to_list()
logn_aft_pd
```

```python
loglog_aft = LogLogisticAFTFitter()
loglog_aft.fit(survival, duration_col='tenure', event_col='churn_Yes')
loglog_aft_pd = loglog_aft.predict_survival_function(survival).transpose()
avg_loglog = loglog_aft_pd.mean().to_list() #computing the average survival time for each month
loglog_aft_pd
```

```python
plt.plot(avg_loglog, color = 'blue', label = 'LogLogistic')
plt.plot(avg_we, color = "green", label = "Weibull")
plt.plot(avg_logn, color = "red", label = "LogNormal")
plt.legend()
plt.title("All the curves")
plt.show()
```

```python

for model in [wb, ex, log, loglogis]:
    model.fit(durations = survival["tenure"], event_observed = survival["churn_Yes"])
    print("The AIC value for", model.__class__.__name__, "is",  model.AIC_)
```

AIC values serve as a metric to assess model fit, with lower values indicating better fit. 
The WeibullFitter model has an AIC of 3216.861, suggesting a moderate fit.
The ExponentialFitter model slightly outperforms it with an AIC of 3215.961, indicating a marginally better fit.
However, the LogNormalFitter model stands out with the lowest AIC of 3209.035, indicating the best fit among the models discussed.
Lastly, the LogLogisticFitter model trails with an AIC of 3214.415.

```python

logn_aft.print_summary()  
```

The interpretation of the model results suggests that for every one unit increase in the independent variable, we expect a (exp(coef) - 1)*100% increase in the dependent variable (churn_Yes), while holding all other variables constant.

For instance, considering the 'age' variable, an exp(coef) of 1.06 implies that with each one-year increase in age, the likelihood of churn occurrence increases by (1.06 - 1)*100%, keeping other variables constant.

To determine significant features for further processing, we typically look at p-values. If a predictor variable has a high p-value, typically above a certain threshold like 0.05, it's considered non-significant and can be removed from the model.

In this case, 'forward_Yes', 'income', 'region_Zone2', and 'region_Zone3' have p-values of 0.35, 0.2, 0.95, and 0.47 respectively. These values exceed the typical significance threshold of 0.05, indicating non-significance. Therefore, these variables are removed, retaining only the significant features for the final model.

```python
survival= survival[['age','churn_Yes','tenure','marital_Married','internet_Yes', 'custcat_E-service',\
                          'custcat_Plus service','custcat_Total service']]
```

```python
logn_aft = LogNormalAFTFitter()
logn_aft.fit(survival, duration_col='tenure', event_col='churn_Yes')
logn_aft_pd = logn_aft.predict_survival_function(survival).transpose()
logn_aft_pd
```

```python
logn_aft.predict_hazard(survival).transpose()
```

```python
logn_aft.print_summary() 
```

```python
MM = 1300
sequence = range(1,len(logn_aft_pd.columns)+1)
r = 0.1

for num in sequence:
    logn_aft_pd.loc[:,num] = logn_aft_pd.loc[:,num]/((1+r/12)**(sequence[num-1]-1))
```

```python
logn_aft_pd
```

```python
logn_aft_pd["CLV"] = MM * logn_aft_pd.sum(axis = 1)
logn_aft_pd
```

```python
logn_aft_pd["CLV"] = MM * logn_aft_pd.sum(axis = 1)
logn_aft_pd
```

```python
telco["CLV"] = logn_aft_pd.CLV
```

The analysis indicates several factors influencing churn risk within this dataset. According to the log-normal accelerated failure time model, customer age, customer category, internet availability, marital status, and region of residence emerge as significant predictors of churn risk. Specifically, younger customers, those with e-service or total service, individuals lacking internet access, unmarried customers, and residents in Zone 2 exhibit higher churn risk. Conversely, variables such as region_Zone 3, income, and forward_Yes demonstrate no significant impact on churn risk.

Regarding profitability, the most lucrative customer segments, based on their average Customer Lifetime Value (CLV), are retirees, followed by those in the Plus Service customer category, and individuals with incomplete high school education. Consequently, these segments are deemed valuable due to their potential to yield higher profits.
